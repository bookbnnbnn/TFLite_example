```python
# 引入套件
import tkinter as tk
from tkinter import ttk

# 建立應用程式主視窗
window = tk.Tk()
# 標題
window.title('猜數字 App')
# 視窗大小
window.geometry('400x200')
# 視窗背景顏色
window.configure(background='white')
# 說明欄、文字區塊
text = '猜數字小遊戲\n\n分成簡單&困難模式\n簡單:沒有限制\n困難:限制5次\n'
header_label = tk.Label(window, text=text)
header_label.pack()


# 難易度分級(hard_frame 群組)
hard_frame = tk.Frame(window)
hard_frame.pack(side=tk.TOP) # 向上對齊
hard_label = ttk.Label(hard_frame, text = '難易度')
hard_label.pack(side=tk.LEFT) # 向左對齊
# 設置下拉選單
cv = tk.StringVar()
comboxlist = ttk.Combobox(hard_frame, textvariable = cv)
comboxlist.pack(side=tk.TOP)
comboxlist["value"]=("簡單","困難") # 設成簡單和困難兩種模式
comboxlist.current(0)  #選擇第一個  
comboxlist.pack() 

# 隨機製造四個數字為答案
def make_ans():  
    import numpy as np
    L = list('0123456789')
    ans = np.random.choice(L, 4, replace = False)
    ans = "".join(ans)
    return ans

# 當按下開始遊戲，跳出另一個視窗
def create():
    ans = make_ans() # 製造答案
    mode = str(comboxlist.get()) # 抓取模式為何
    global times
    times = 1000 if mode == '簡單' else 5 # 若是簡單則可遊玩次數設為1000(設一個相對5的極大值)，若是困難則設為5
    guess_list = ['' for i in range(times)] # 設一個空的list用來儲存結果
    # 開始遊戲
    def play():
        # 比對猜的數字和答案
        guess = str(guess_entry.get())
        def guess_number():
            a, b = 0, 0
            for i in range(len(ans)):
                # 數字對且位置對
                if ans[i] == guess[i]:
                    a += 1
                # 數字對但位置錯
                elif ans[i] in guess:
                    b += 1
            res = '答對了' if a == 4 else f'{a}A{b}B'
            return res
        # 檢查 
        def check():
            flag = False
            # 輸入不是文字的東西
            if guess.isdigit() == False: 
                res = '請輸入數字'
                flag = True
            # 輸入不是四位數字
            elif len(guess) != 4:
                res = '請輸入4位數字'
                flag = True
            #輸入重複的數字
            elif sum([guess.count(i) for i in guess]) > 4: 
                res = '請不要輸入重複的數字'
                flag = True
            # 若檢查都通過則比對數字
            else: 
                res = guess_number()
                flag = False
            return flag, res
        
        global times
        condition = check()
        # 若檢查有錯誤顯示錯誤的原因
        if condition[0]:
            text = ''.join(guess_list)
            result_label.configure(text = text + condition[1])
        # 若檢查沒有錯誤顯示比對答案的結果
        elif condition[0] == False:
            times -= 1  
            guess_list[times] = f'你的答案{guess}，{condition[1]}\n' # 將本次猜測的數字放入guess_list中
            text = ''.join(guess_list) # 將guess_list中的元素轉成字串並合併
            # 若是簡單模式則不顯示次數
            if comboxlist.get() == '簡單':
                text = f'失敗，答案是{ans}' if times <= 0 else text
                result_label.configure(text = text)
            # 若是困難模式則顯示次數
            elif comboxlist.get() == '困難':
                text = f'失敗，答案是{ans}' if times <= 0 else text + f'剩餘{times}次'
                result_label.configure(text =text)
    def give_up():
        global times 
        times = 0
        text = f'失敗，答案是{ans}'
        result_label.configure(text =text)

    # 跳出另一個視窗
    window2 = tk.Toplevel()
    window2.title('開始遊戲')
    window2.geometry('450x400')
    window2.configure(background='white')

    # 以下為 guess_frame 群組
    guess_frame = tk.Frame(window2)
    guess_frame.pack(side=tk.TOP)
    guess = tk.StringVar()
    guess_label = tk.Label(guess_frame, text = '你猜的數字')
    guess_label.pack(side=tk.LEFT)
    guess_entry = tk.Entry(guess_frame, textvariable = guess)
    guess_entry.pack(side=tk.LEFT)
    
    result_label = tk.Label(window2)
    result_label.pack()

    tk.Button(window2, text='猜測', command=play).pack()
    tk.Button(window2, text='放棄', command=give_up).pack()

tk.Button(window, text='開始遊戲', command=create).pack()
window.mainloop()
```
